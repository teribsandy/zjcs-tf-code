分词：要求分解单词列表、单词频次字典、按单词频次逆序输出结果

```python
with open('罗密欧与朱丽叶(英文版)莎士比亚.txt','r',encoding='gbk') as f:
    content=f.read()
print (content)
```

```python
word_list=content.split()
print (word_list)
```

```python
import string
```

```python
word_dict={}
for each in word_list:
    each=each.strip(string.punctuation)
    if each not in word_dict.keys():
        word_dict[each]=1
    else:
        word_dict[each]+=1
print (word_dict)
```

```python
word_order=sorted(word_dict.items(),key=lambda x: x[1],reverse=True)
for each in word_order:
    print (each[1],each[0])
```
